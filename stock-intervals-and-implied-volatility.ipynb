# Task 3(Bonus): Calculate the Implied Volatility

```python
import numpy as np
import pandas as pd
from scipy.stats import norm
from scipy.optimize import minimize_scalar
```

```python
def bs_call(S, K, T, r, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    return S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
def bs_put(S, K, T, r, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    return K * np.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)

def error_function(sigma, S, K, T, r, market_price):
    
    bs_price = bs_call(S, K, T, r, sigma)
    return abs(market_price - bs_price)

def error_functions(sigma, S, K, T, r, market_price):
    
    bs_price = bs_put(S, K, T, r, sigma)
    return abs(market_price - bs_price)
```

```python
df = df.reset_index(drop=True)
df
```

```python

for index,row in df.iterrows():
 
    S = row['Close']
    K = row['Strike_Price']
    T = row['y_remain']
    r = 0.05  # risk-free interest rate
    market_price = row['Close']
    
    if row['Flag']=="CE":
        implied_vol = minimize_scalar(error_function, args=(S, K, T, r, market_price),
                                      bounds=(0, 2), method='bounded').x
        df.at[index, 'Implied_vol'] = implied_vol
    else:
        implied_vol = minimize_scalar(error_functions, args=(S, K, T, r, market_price),
                                      bounds=(0, 2), method='bounded').x
        df.at[index, 'Implied_vol'] = implied_vol       
```
